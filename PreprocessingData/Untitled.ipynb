```python
import pandas as pd
import os
```

```python
OCCUPATIONS = { 0: "other-or-ot-pecified", 1: "academic/educator", 2: "artist", 3: "clerical/admin",
                4: "college/grad-tudent", 5: "customer service", 6: "doctor/health care",
                7: "executive/managerial", 8: "farmer", 9: "homemaker", 10: "K-12-student", 11: "lawyer",
                12: "programmer", 13: "retired", 14: "sales/marketing", 15: "scientist", 16: "self-employed",
                17: "technician/engineer", 18: "tradesman/craftsman", 19: "unemployed", 20: "writer" }
```

```python
users=pd.read_csv('users.dat',sep='::',
                  engine='python',names=['user_id','gender','age','occupation_id','zipcode'])
```

```python
users['occupation']=users['occupation_id'].apply(lambda x:OCCUPATIONS[x])
```

```python
os.chdir('..')
os.chdir('Data')
users.to_csv('user.csv',sep=',',header=True,encoding='latin-1',columns=['user_id','gender','age','occupation_id','zipcode','occupation'])
os.chdir('..')
os.chdir('PreprocessingData')
```
